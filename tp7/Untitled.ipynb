```python
import cv2 as cv
import numpy as np

img = cv.imread('nSpeyoLESmG6w-u2amRdww.jpg', cv.IMREAD_COLOR)

h, w, c = img.shape

img_gray = np.zeros((h, w), dtype=np.uint8)

for y in range(h):
    for x in range(w):
        b, g, r = img[y, x]
        gray_value = (b * 0.3333 + g * 0.3333 + r * 0.3333) 
        img_gray[y, x] = np.uint8(gray_value) 

cv.imshow('Original Image', img)
cv.imshow('Grayscale Image (Manual Mean)', img_gray)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np

img = cv.imread('nSpeyoLESmG6w-u2amRdww.jpg', cv.IMREAD_COLOR)

h, w, c = img.shape

img_h = np.zeros((h, w), dtype=np.uint8)
img_s = np.zeros((h, w), dtype=np.uint8)
img_v = np.zeros((h, w), dtype=np.uint8)

for y in range(h):
    for x in range(w):
        b, g, r = img[y, x]

        r = r / 255.0
        g = g / 255.0
        b = b / 255.0

        mx = max(r, g, b)
        mn = min(r, g, b)
        df = mx - mn

        if df == 0:
            h = 0
        elif mx == r:
            h = (60 * ((g - b) / df) + 360) % 360
        elif mx == g:
            h = (60 * ((b - r) / df) + 120) % 360
        elif mx == b:
            h = (60 * ((r - g) / df) + 240) % 360

        if mx == 0:
            s = 0
        else:
            s = (df / mx) * 100

        v = mx * 100

        img_h[y, x] = np.uint8(h // 2)
        img_s[y, x] = np.uint8(s)
        img_v[y, x] = np.uint8(v)

cv.imshow('Original Image', img)
cv.imshow('Hue Channel', img_h)
cv.imshow('Saturation Channel', img_s)
cv.imshow('Value Channel', img_v)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2 as cv
import numpy as np

img = cv.imread('nSpeyoLESmG6w-u2amRdww.jpg', cv.IMREAD_COLOR)

h, w, c = img.shape

img_hsv = np.zeros((h, w, 3), dtype=np.uint8)

for y in range(h):
    for x in range(w):
        b, g, r = img[y, x]

        r = r / 255.0
        g = g / 255.0
        b = b / 255.0

        mx = max(r, g, b)
        mn = min(r, g, b)
        df = mx - mn

        if df == 0:
            h = 0
        elif mx == r:
            h = (60 * ((g - b) / df) + 360) % 360
        elif mx == g:
            h = (60 * ((b - r) / df) + 120) % 360
        elif mx == b:
            h = (60 * ((r - g) / df) + 240) % 360

        if mx == 0:
            s = 0
        else:
            s = (df / mx) * 100

        v = mx * 100

        img_hsv[y, x] = [np.uint8(h // 2), np.uint8(s), np.uint8(v)]

cv.imshow('Original Image', img)
cv.imshow('HSV Image', img_hsv)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Initialize global variables
th = 0
type_th = 0
adaptive_type = 0  
color_map = 0      
imgRes_color = None  # Initialize this variable to use it later

# Load the grayscale image
img = cv2.imread('nSpeyoLESmG6w-u2amRdww.jpg', cv2.IMREAD_GRAYSCALE)
if img is None:
    print("Error: Image not found.")
    exit()

def afficher():
    global img, th, type_th, adaptive_type, color_map, imgRes_color
    
    imgRes = np.zeros_like(img)

    if adaptive_type == 0:
        sup_th = img > th
        inf_th = np.invert(sup_th)
        
        if type_th == 0:
            imgRes[sup_th] = 255
            imgRes[inf_th] = 0
        elif type_th == 1:
            imgRes[sup_th] = 0
            imgRes[inf_th] = 255
        elif type_th == 2:
            imgRes[sup_th] = th
            imgRes[inf_th] = img[inf_th]
        elif type_th == 3:
            imgRes[sup_th] = img[sup_th]
            imgRes[inf_th] = 0
        elif type_th == 4:
            imgRes[sup_th] = 0
            imgRes[inf_th] = img[inf_th]
    else:
        if adaptive_type == 1:
            imgRes = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                           cv2.THRESH_BINARY, 11, 2)
        elif adaptive_type == 2:
            imgRes = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                           cv2.THRESH_BINARY, 11, 2)
    
    # Convert original image to color
    img_color = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    
    # Apply color map if selected
    if color_map > 0:
        imgRes_color = cv2.applyColorMap(cv2.cvtColor(imgRes, cv2.COLOR_GRAY2BGR), color_map - 1)
    else:
        imgRes_color = cv2.cvtColor(imgRes, cv2.COLOR_GRAY2BGR)

    # Combine the images for display
    combined_view = np.hstack((img_color, imgRes_color))
    
    cv2.imshow("Original and Thresholded", combined_view)

def change_th(x):
    global th
    th = x
    afficher()

def change_type(x):
    global type_th
    type_th = x
    afficher()

def change_adaptive_type(x):
    global adaptive_type
    adaptive_type = x
    afficher()

def change_color_map(x):
    global color_map
    color_map = x
    afficher()

# Initialize the display
afficher()

# Create trackbars for user interaction
cv2.createTrackbar("Threshold", "Original and Thresholded", 0, 255, change_th)
cv2.createTrackbar("Type", "Original and Thresholded", 0, 4, change_type)
cv2.createTrackbar("Adaptive", "Original and Thresholded", 0, 2, change_adaptive_type)
cv2.createTrackbar("ColorMap", "Original and Thresholded", 0, 11, change_color_map)

while True:
    key = cv2.waitKey(1) & 0xFF
    if key == 27:  # Escape key
        print("Exiting...")
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np


img = cv2.imread('nSpeyoLESmG6w-u2amRdww.jpg', cv2.IMREAD_GRAYSCALE)
if img is None:
    print("Error: Image not found.")
    exit()

def calculate_gradient_magnitude(image):
    # Calculate the gradients in the x and y directions
    grad_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Gradient in x
    grad_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Gradient in y

    # Calculate the magnitude of the gradient
    grad_magnitude = np.sqrt(grad_x**2 + grad_y**2)

    # Normalize the gradient magnitude to the range [0, 255]
    grad_magnitude = cv2.normalize(grad_magnitude, None, 0, 255, cv2.NORM_MINMAX)
    
    return grad_magnitude.astype(np.uint8)  # Convert to uint8 type

def afficher():
    global img
    
    img_grad_magnitude = calculate_gradient_magnitude(img)

    # Convert the original grayscale image to color
    img_color = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

    # Ensure img_grad_magnitude is 3-channel for stacking
    img_grad_magnitude_color = cv2.cvtColor(img_grad_magnitude, cv2.COLOR_GRAY2BGR)

    # Combine the images for display
    combined_view = np.hstack((img_color, img_grad_magnitude_color))
    
    cv2.imshow("Original and Gradient Magnitude", combined_view)


afficher()

while True:
    key = cv2.waitKey(1) & 0xFF
    if key == 27:  # Escape key
        break

cv2.destroyAllWindows()
```
